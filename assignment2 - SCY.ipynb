### 1번 문제

* 다음 6 개의 Expression에 대해 Evaluation 결과 값을 출력하고, 해당 결과가 나온 이유에 대해 설명하시오
    * 1 and 2 and 3 and 4
    * 1 or 2 or 3 or 4
    * 1 and 2 or 3 and 4
    * (1 and 2) or (3 and 4)
    * 1 or 2 and 3 or 4
    * (1 or 2) and (3 or 4)

```python
print (1 and 2 and 3 and 4)
print (1 or 2 or 3 or 4)
print (1 and 2 or 3 and 4)
print ((1 and 2) or (3 and 4))
print (1 or 2 and 3 or 4)
print ((1 or 2) and (3 or 4))
```

* 1 and 2 and 3 and 4
    * and연산자를 사용할 때 앞의 값이 true이면 뒤의 값도 확인을 해 주어야 한다. 여기서는 연산이 끝날때까지 true값만 나오므로 마지막의 true값이 출력된다. 그래서 4가 출력되는 것이다.
* 1 or 2 or 3 or 4
    * or연산자를 사용할 때 앞의 값이 true이면 뒤의 값은 확인을 할 필요가 없다. or연산자를 사용할 때는 하나만 true이면 결과가 true이기 때문이다. 여기서는 처음부터 true값인 1이 나오므로 1이 출력된다.
* 1 and 2 or 3 and 4
    * and가 or보다 우선순위가 높으므로 1 and 2와 3 and 4가 먼저 계산된다. 1 and 2는 2고 3 and 4는 4인데 이들을 or연산하면 처음에 true값으로 2가 나왔으므로 뒤는 볼 필요도 없이 true값인 2가 출력된다.
* (1 and 2) or (3 and 4)
    * and가 or보다 우선선위가 높으므로 여기의 괄호는 의미가 없다. 바로 위의 문제와 같은 과정을 거쳐 같은 값이 출력된다.
* 1 or 2 and 3 or 4
    * and가 or보다 우선순위가 높으므로 2 and 3이 먼저 계산된다. 2 and 3은 마지막 true값인 3이고, 이를 가지고 1 or 3을 하면 처음에 나온 true값인 1이 된다. 그리고 이걸로 1 or 4를 하면 처음에 나온 true값인 4가 출력된다.
* (1 or 2) and (3 or 4)
    * 괄호가 쳐진 것들이 우선순위를 가지므로 1 or 2와 3 or 4가 먼저 계산된다. 1 or 2는 1이고 3 or 4는 3인데 이들을 and연산하면 둘 다 true값이므로 마지막에 나온 true값인 3이 출력된다.

### 2번 문제

* 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
    * 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
    * 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
    * 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
    * 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
        * div_path('/usr/local/bin/python')
            * ['/usr/local/bin', 'python']
        * div_path('/home/chulsoo/test.txt')
            * ['/home/chulsoo', 'test.txt']
* [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다. 

```python
def div_path(s):
    l = s.rsplit('/', 1)
    return l
print(div_path('/usr/local/bin/python'))
```

경로에 해당하는 문자열의 원소 중 가장 오른쪽에 있는 /가 디렉토리 경로명과 파일명을 구분하는 /다. 따라서 문자열 메소드인 rsplit을 사용해 문자열 뒤에서부터 문자열을 훑으며 /가 나오면 분리하여 리스트에 반환한다.

### 3번 문제

* 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
    * 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
    * 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
    * 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
    * 반환하는 리스트는 정렬이 되어 있어야 한다.
    * 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
        * list_union([1, 2, 3], [1, 2, 4])
            * [1, 2, 3, 4]
        * list_union([-10, -5, 0, -1], [100, 9, 0, 9])
            * [-10, -5, -1, 0, 9, 100]
        * list_union([0, 1, 2], [0, 1, 2])
            * [0, 1, 2]
    * [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
    * [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb):
    res = []
    for x in lista + listb:
        if x not in res:
            res.append(x)
    res.sort()
    return res
    
print(list_union([1,2,3], [1,2,4]))
print(list_union([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_union([0, 1, 2], [0, 1, 2]))
```

두 리스트를 더하는데 이 문제에서 요구하는 것은 두 리스트의 합집합을 구하는 것이므로 중복된 원소가 있으면 안된다. 그래서 조건문을 사용하여 두 리스트를 더하면서 합집합 리스트에 들어가는 원소가 이미 합집합 리스트에 있는 원소 중 동일한 것이 없으면 이를 리스트 메소드인 append를 사용해 합집합 리스트에 추가하도록 한다. 반환하는 리스트는 정렬이 되어 있어야 하므로 리스트 메소드인 sort를 사용하여 정렬한다. 이렇게 하면 이 문제를 풀 수 있다. 

### ACM-ICPC 문제 1110번

* 0보다 크거나 같고, 99보다 작거나 같은 정수가 주어질 때 다음과 같은 연산을 할 수 있다. 먼저 주어진 수가 10보다 작다면 앞에 0을 붙여 두 자리 수로 만들고, 각 자리의 숫자를 더한다. 그 다음, 주어진 수의 가장 오른쪽 자리 숫자와 앞에서 구한 합의 가장 오른쪽 자리 숫자를 이어 붙이면 새로운 수를 만들 수 있다. 다음 예를 보자. 26부터 시작한다. 2+6 = 8이다. 새로운 숫자는 68이다. 6+8 = 14이다. 새로운 숫자는 84이다. 8+4 = 12이다. 새로운 숫자는 42이다. 4+2 = 6이다. 새로운 숫자는 26이다. 위의 예는 4번만에 원래 숫자로 돌아올 수 있다. 따라서 26의 사이클의 길이는 4이다. N이 주어졌을 때, N의 사이클의 길이를 구하는 프로그램을 작성하시오.

    * 입력 : 첫째 줄에 N이 주어진다. N은 0보다 크거나 같고, 99보다 작거나 같은 정수이다.
    * 출력 : 첫째 줄에 N의 사이클 길이를 출력한다.

```python
n = int(input())

add = n
cnt = 0 

while True:
    if add < 10:
        add = (add*10) + add
    else:
        r = int(add / 10) + (add % 10)
        add = (add % 10 * 10) + (r % 10)
    cnt += 1
    if add is n:
        break

print(cnt)
```

사용자로부터 정수를 하나 입력받는다. 먼저 이 수가 10보다 작은가를 판별한다. 그렇다면 앞에 0을 붙여 두 자리 수로 만들어야 하기 때문이다. 만약 10보다 작지 않다면 바로 int(add / 10) + (add % 10)를 하여 각 자리의 수를 더한다. 이 결과값의 오른쪽 수(r % 10)와 합의 오른쪽 숫자(add % 10 * 10)를 더한 값을 변수 r에 할당한다. 그리고 문제에서 요구하는 정수의 사이클의 길이를 구하기 위해 카운트(cnt)를 1 증가시킨다. 이 과정을 반복문 내에서 거치다가 만약 사용자로부터 입력받은 정수와 add의 값이 일치하게 되면 반복문을 빠져나온다. 이렇게 하면 이 문제를 풀 수 있다. 

### [Incremental Project] 

* [Incremental Project] 웹 URL로 지정된 웹페이지를 문자열로 가져와 1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.
    * HTML 태그는 다음 조건을 가지는 것이다.
        * < 로 시작하여 >로 끝난다.
        * 예를 들어, 아래와 같은 HTML 내용 내에 한 줄이 있다고 가정하자.
              <label for="keep_signed">로그인 유지</label>
        * 위 소스에서 HTML 태그를 제외한 순수 텍스트 문자열은 아래와 같다.
              로그인 유지
        * 한편, 아래와 같은 HTML 한 줄이 있다고 가정하자.
              <link rel="stylesheet" href="/common/css/xe.min.css?20150910092654" />
        * 위 소스는 HTML 태그로만 구성된 것이며 걸러낼 수 있는 순수 텍스 문자열은 존재하지 않는다.
    * 단어를 나누는 기준은 오로지 공백문자 (whitespace)이다.
        * 즉, 위 순수 텍스트 문자열인 "로그인 유지"에서 분리한 단어는 "로그인"과 "유지"이다.
    * 중복을 고려하지 않고 무조건 모든 단어의 개수를 세어서 출력한다.
    * 본 숙제를 위하여 HTML 파싱과 관련된 모듈 (예. BeautifulSoup) 사용 금지
    * 다음은 URL로 지정된 웹페이지를 특정 문자열로 가져오는 코드이다.
        * 즉, 숙제 코드는 아래 코드로 부터 시작해야 한다.

```python
from urllib import request 
 
url = 'http://cse.koreatech.ac.kr'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')

# 추후 숙제 답안 제출 때엔 아래 코드는 주석처리할 것                 
#print(type(source_bytes))
#print(type(source))
#print(source)
#print(source)

word_flag = True
fst = 0
lst = 0
cnt = 0
l = []
flag = True

for x in source:
    if(x == '<'):
        word_flag = False
    elif(x == '>'):
        word_flag = True
        
    if(flag == True and word_flag == True and x != ' ' and x != '>' and x != '\n' and x != '\t'):
        flag = False
        fst = cnt
    elif(flag == False and (x == ' ' or x == '<') and x != '\n' and x != '\t'):
        flag = True
        lst = cnt
        l.append(source[fst:lst])
        
    cnt=cnt+1

print(l)
print("단어의 총 개수 : "+str(len(l)))
```

단어를 걸러내기 위해 우선 HTML 태그의 시작인 <가 나오면 HTML 태그의 끝인 >가 나올 때까지 그 사이에 있는 것들은 아예 검사를 하지 않도록 했다. 그 외에는 공백을 기준으로 단어인지를 검사하여 리스트에 저장할 수 있도록 하였다.(문자를 검사할 때마다 카운트(cnt)를 늘려서 그 문자의 위치를 저장하여 리스트 메소드인 append를 사용해 단어 리스트에 추가함) 이렇게 문제를 푼다고 풀었지만 사실 정확히 풀이한 것은 아니다. 아직도 이유를 잘 모르겠지만 분명히 \n와 \t을 걸렀는데도 단어 리스트에 존재하고, 심지어는 있어서는 안 될 >까지 있다. 하루빨리 교수님께서 문제를 설명해 주셨으면 한다.
